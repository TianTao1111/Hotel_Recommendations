```python
df["distance"] = df["distance"].apply(lambda x: float(x.split(" ")[0]) if x.split(" ")[1]=='km' 
                                      else float(x.split(" ")[0])/1000 if x.split(" ")[1]=='m'
                                      else None)
```

```python
# a function to transfer distance
def lam(x):
    new = x.split(" ")
    if new[1]  == 'km':
        y = float(new[0])
    elif new[1] == 'm':
        y = float(new[0])/1000

    return y

x="250 m from centre"
lam(x)
```
